```python
database = {}
for i in list(os.listdir()):
    if '.csv' in i:
        key = (i[:i.find('.csv')])
        database[key] = pd.read_csv(i)
        
```

```python
conn = sqlite3.connect('weather.db')
for i in database.keys():
    database[i].to_sql(i, conn)
```
